```python
# Catagory the people by their age like children, adult, citizen, senior citizen...
# age: 20
# Adult
#-----------
#children <18, adult <30, citizen < 60, senior citizen >60 
```

```python
age= int(input("age: "))
if(age < 18):
    print("Children")
elif(age < 30):
    print("Adult")
elif(age < 60):
    print("Citizen")
elif(age >= 60):
    print("Senior Citizen")
else:
    print("Please enter a proper number")
```
