```python
#1.Find the GCD of Two Numbers
#Question: Write a Python program to find the Greatest Common Divisor (GCD) of 36 and 60.


def gcd_iter(a, b):
    while b != 0:
        # a becomes b, and b becomes the remainder of a % b
        a, b = b, a % b
    return a

# Example Usage
num1=36
num2=60
print(f"The GCD is: {gcd_iter(num2, num1)}") 
```

```python
#3. Find All Armstrong Numbers Between 1 and 1000
#Question: Write a Python program to print all Armstrong numbers between 1 and 1000.
print("Armstrong numbers between 1 and 1000:")
print("-------------------------------------")
for num in range(1, 1001):
    # change number to string to get length and individual digits
    s_num = str(num)
    number_of_digits = len(s_num)
    
    # compute the sum of digits raised to the power of num_digits
    sum_of_digit_raised_to_power = sum(int(digit) ** number_of_digits for digit in s_num)
    
    # if the sum equals the original number
    if sum_of_digit_raised_to_power == num:
        print(num)
```

```python
#4.Count the Frequency of Each Character in a String
#Question: Write a Python program to count the frequency of each character in "programming".

word = "programming"
freq_of_char = {}

for char in word:
    if char in freq_of_char:
        freq_of_char[char] += 1
    else:
        freq_of_char[char] = 1

print("Character frequencies:")
for char, count in freq_of_char.items():
    print(f"'{char}': {count}")
```
